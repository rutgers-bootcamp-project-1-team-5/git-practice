```python
import json
import requests
from api_keys import marketaux_key
```

```python
params = {
    'symbols' : 'META',
    #'countries' : 'us',
    'filter_entities' : 'true',
    'api_token' : marketaux_key
}

response = requests.get('https://api.marketaux.com/v1/news/all', params=params)
```

```python
print(json.dumps(response.json(), indent=4))
```

```python
stop_words = ['<em>', '</em>', '.', ',', '“', '”', '\"', '(', ')', '\n']
word_count = {}
keys_to_drop = ['in', 'the', 'of', 'on', 'which', 'its', 'and', 'that', 'a', 'with', '-', 'was', 'as']
article_data = response.json()
```

```python
for i in article_data['data']:
    for j in i['entities']:
        for k in j['highlights']:
            highlight = k['highlight'].lower()
            for r in stop_words:
                highlight = highlight.replace(r, '')
            for word in highlight.split(' '):
                try:
                    word_count[word] += 1
                except KeyError:
                    word_count[word] = 1

for key in keys_to_drop:
    word_count.pop(key, None)

word_count_sorted = sorted(word_count.items(), key=lambda item:item[1], reverse=True)

for key, value in word_count_sorted:
    print(f'{key}: {value}')
```
