```python
import json
import requests
from pathlib import Path
import pandas as pd

list_articles = ['articles_META_20211109-20220118.json','articles_META_20220120-20220203.json', 'articles_META_20220909-20220915.json']

stop_words = ['<em>', '</em>', '.', ',', '“', '”', '\"', '(', ')', '\n']

words_pd = pd.DataFrame()


# Word counts that we don't care about
keys_to_drop = ['in', 'the', 'of', 'on', 'which', 'its', 'and', 'that', 'a', 'with', '-', 'was', 'as', 'to', 'for', 'it', 'is', 'from', 'this']
```

```python
index = 0
for article_temp in list_articles:
    
    file_path = Path('articles', article_temp)
    json_file = open(file_path, 'r', encoding='utf-8')
    json_data = json.load(json_file)

    word_count = {}

    for grouping in json_data:
        for data in grouping['data']:
            for entity in data['entities']:
                for highlight in entity['highlights']:
                    # Set text to lowercase to prevent duplicate word counts (i.e. META and Meta)
                    text = highlight['highlight'].lower()
                    # Remove stop words
                    for word in stop_words:
                        text = text.replace(word, '')
                    # Count words
                    for word in text.split(' '):
                        try:
                            word_count[word] += 1
                        except KeyError:
                            word_count[word] = 1

    # Get rid of uninteresting words
    for key in keys_to_drop:
        word_count.pop(key, None)

    word_count_sorted = sorted(word_count.items(), key=lambda item:item[1], reverse=True)

    print(word_count_sorted)

    title = f'Day {index}'


    words_pd_temp = pd.DataFrame(word_count_sorted, columns= ['Words', title])
    words_pd_temp.head()

    if index!=0:
        words_pd = words_pd.merge(words_pd_temp, on='Words', how = 'outer')
    else:
        words_pd= words_pd_temp

    index+=1

words_pd
```
